```python
import pandas

df = pandas.read_csv( "/content/sample_data/CarPrice_Assignment.csv"  )

df
```

```python
y = df[ "price"]

y
```

```python
x = df[[ "symboling",	"wheelbase",	"carlength",	"carwidth",	"carheight",	"curbweight",	"enginesize",	"boreratio",	"stroke",	"compressionratio",	"horsepower",	"peakrpm",	"citympg",	"highwaympg"]]

x
```

```python
116 * 0.8

y_train = y [  0 : 93 ]

y_train
```

```python
x_train = x [ 0:93 ]

x_train
```

```python
y_test = y[93:  ]

y_test
```

```python
x_test = x[93: ]
x_test
```

```python
from keras.models import Sequential

from keras.layers import Dense, Activation

B_model = Sequential()

B_model.add( Dense( 32, activation = "relu", input_dim = 14   )    )

B_model.add( Dense( 32, activation = "relu"   )    )

B_model.add( Dense( 32, activation = "relu"   )    )

B_model.add( Dense( 1 ) )

B_model.compile(loss = "mean_squared_error", optimizer = "adam")
```

```python
B_model.fit(x_train, y_train, epochs = 3000)
```

```python
x_test

y_test

y_predicted = B_model.predict(x_test)

y_predicted
```

```python
## reshaping
y_predicted_1 = y_predicted.reshape(1,112)

import numpy
y_test_1 = y_test.to_numpy()

APE = abs( numpy.subtract(y_predicted_1, y_test_1) )/y_test_1*100

APE
```

```python
MAPE = numpy.mean(APE)
MAPE
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# Load dataset (adjust the file path as needed)
df = pd.read_excel("/content/sample_data/Midterm-Dataset_2 (1).xlsx")
df = df.rename(columns=lambda x: x.strip())
df = df.drop(columns=["Date"])  # Drop non-numerical column
print(df.columns.tolist())
```

```python
df = pd.get_dummies(df, columns=["Location", "living Quality"], drop_first=False)
df = df.astype(int)
```

```python
# Define features and target variable
X = df.drop(columns=["Consumption"])
X
```

```python
y = df["Consumption"]
y
```

```python
# Split dataset into training and testing parts
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Build the Neural Network model
model = Sequential()
# Input layer: number of neurons = 32 with ReLU activation; input_dim set to number of features
model.add(Dense(32, activation='relu', input_dim=X.shape[1]))
# Additional hidden layers
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
# Output layer: one neuron (for regression)
model.add(Dense(1))

# Compile the model using mean squared error loss and the adam optimizer
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model (adjust epochs for your needs; here we use 3000 epochs)
model.fit(x_train, y_train, epochs=3000, verbose=1)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
y_pred
```

```python
# Calculate Mean Absolute Percentage Error (MAPE)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"MAPE: {mape * 100}%")
```

```python
# ---- Prediction for a specific input ----
# We need to predict the consumption when Income = $33,000.
# For Job type, we'll take the mode (most common value).
mean_job_type = int(X["Job type"].mode()[0])
# For dummy variables, we use their mean values (even though they are binary, the average can serve as a proxy).
other_means = X.drop(columns=["Income", "Job type"]).mean().to_list()
# Build the input vector; note that the order must match X's columns.
# Here X is ordered as: ['Income', 'Job type', <dummy columns...>]
input_vector = [33000, mean_job_type] + other_means

input_array = np.array([input_vector])  # Convert to 2D array
predicted_33000 = model.predict(input_array)[0][0]
print(f"Predicted consumption at $33,000 income: ${predicted_33000:,.2f}")
```

```python
# ---- Simulate the impact of varying income ----
# Simulate incomes from $25,000 to $45,000 at steps of $1,000; other features remain constant.
income_range = np.arange(25000, 46000, 1000)
simulated_predictions = []

for income in income_range:
    vector = [income, mean_job_type] + other_means
    pred_value = model.predict(np.array([vector]))[0][0]

    simulated_predictions.append(pred_value)
```

```python
# Display simulation results
print("\nSimulated Income vs Predicted Consumption:")
for income, pred in zip(income_range, simulated_predictions):
    print(f"Income: ${income} → Predicted Consumption: ${pred:,.2f}")
```
